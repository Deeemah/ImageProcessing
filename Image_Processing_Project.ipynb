```python
#Import required libraries

from rembg import remove
import matplotlib.pyplot as plt 
import cv2
from google.colab import files
```

```python
print("we have a variety of pictures you can remove the background of,\n Pick an image from 1 to 10 to try one:")

#Reading images
firstimage= cv2.cvtColor(cv2.imread('manselfie.jpg'), cv2.COLOR_BGR2RGB)
secondimage=cv2.cvtColor(cv2.imread('person.jpg'), cv2.COLOR_BGR2RGB)
thirdimage=cv2.cvtColor(cv2.imread('girl.jpeg'), cv2.COLOR_BGR2RGB)
fourthimage=cv2.cvtColor(cv2.imread('girl2.jpeg'), cv2.COLOR_BGR2RGB)
fifthimage=cv2.cvtColor(cv2.imread('elderlady.jpeg'), cv2.COLOR_BGR2RGB)
sixthimage=cv2.cvtColor(cv2.imread('workingwoman.webp'), cv2.COLOR_BGR2RGB)
seventhimage=cv2.cvtColor(cv2.imread('baby.jpeg'), cv2.COLOR_BGR2RGB)
eightthimage=cv2.cvtColor(cv2.imread('boypic.webp'), cv2.COLOR_BGR2RGB)
ninethimage=cv2.cvtColor(cv2.imread('girlpic.jpeg'), cv2.COLOR_BGR2RGB)
tenthimage=cv2.cvtColor(cv2.imread('workingwoman.jpeg'), cv2.COLOR_BGR2RGB)

#appending images
images = []
images.append(firstimage)
images.append(secondimage)
images.append(thirdimage)
images.append(fourthimage)
images.append(fifthimage)
images.append(sixthimage)
images.append(seventhimage)
images.append(eightthimage)
images.append(ninethimage)
images.append(tenthimage)
titles = ['image 1','image 2','image 3','image 4','image 5','image 6','image 7','image 8','image 9','image 10']

#for loop that goes through every images
for i in range(len(images)):
    fig  = plt.figure(figsize=(10,10))
    plt.subplot(4,5,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])


plt.show()
```

```python
choice = int(input("Enter your choice(1-10): "))

output_path = 'output.png' # output image path

#Here we created a menu to let the user choose a picture 
def menu_switch(choice):
    input_path = None
    # switch statement to handle the user's choice
    if choice == 1:
        input_path = 'manselfie.jpg' # input image path
    elif choice == 2:
        input_path = 'person.jpg' # input image path
    elif choice == 3:
        input_path = 'girl.jpeg' # input image path
    elif choice == 4:
        input_path = 'girl2.jpeg' # input image path
    elif choice == 5:
        input_path = 'elderlady.jpeg' # input image path
    elif choice == 6:
        input_path = 'workingwoman.webp' # input image path
    elif choice == 7:
        input_path = 'baby.jpeg' # input image path
    elif choice == 8:
        input_path = 'boypic.webp' # input image path
    elif choice == 9:
        input_path = 'girlpic.jpeg' # input image path
    elif choice == 10:
        input_path = 'workingwoman.jpeg' # input image path
    else:
        print("invalid choice")
    return input_path

#inputImage takes the choice of the user as input to use later
inputImage = cv2.cvtColor(cv2.imread(menu_switch(choice)), cv2.COLOR_BGR2RGB)
```

```python
 #remove background of input image
withouBackGround = remove(inputImage)

#convert the image to grey
grayImage = cv2.cvtColor(inputImage, cv2.COLOR_BGR2GRAY) 

#remove background of gray image
grayWitoutbackGround = remove(grayImage) 


#initializing variables 
choice2 = 0
images = []

#shows original image
images.append(inputImage)

while(choice2!=1 and choice2!=2):
  choice2 = int(input("Do you want the image to be grey or colored? \n(Type 1 for grey and 2 for colored): "))

#if the choice is grey
if choice2==1:
  #append image array
  images.append (grayImage)
  images.append(grayWitoutbackGround)
  titles = ['Original Image','gray Image','Gray Image without Background']
#if the choice is colored  
elif choice2==2:
  #append image array
  images.append(withouBackGround)
  titles = ['Original Image','without background']




#this loop is to go through every image, and imshow is used to display each image
for i in range(len(images)):
    fig  = plt.figure(figsize=(10,10))
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])


plt.show()



```
